# 실습5 에서 출력된 결과

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

#iris = load_iris()

# target을 붙이고 species columns 추가
# df = pd.DataFrame(data = np.c_[iris.data,iris.target], columns = iris.feature_names + ['Species'])
# target에 이름 지정해주기
# df['Species'] = df['Species'].map({0: "setosa",1: "versicolor",2: "virginica"})

df = pd.read_csv('irisData.csv')

grouped = df.groupby('Species')

setosa = grouped.get_group(name='setosa')
versicolor = grouped.get_group(name='versicolor')
virginica = grouped.get_group(name='virginica')

setosa20 = setosa.sample(20)
versic20 = versicolor.sample(20)
virgin20 = virginica.sample(20)

iris20 = setosa20.append(versic20)
iris20 = iris20.append(virgin20)

display(iris20)
```

##  1)3개 종에 대한 boxplot  (subplot 이용) 

```python
import matplotlib.pyplot as plt
```

### Setosa

```python
iris20_grouped = iris20.groupby('Species')

setosa20_group = iris20_grouped.get_group(name='setosa')

setosa20_group.plot.bar(rot=0, subplots=True)
```

### versicolor

```python
versicolor20_group = iris20_grouped.get_group(name='versicolor')

versicolor20_group.plot.bar(rot=0,subplots=True)
```

### virginica

```python
virginica20_group = iris20_grouped.get_group(name='virginica')

virginica20_group.plot.bar(rot=0, subplots=True)
```

## 2)3개 종에 대한 histogram 

```python
def setting(name,x,y):
    plt.title(name)
    plt.xlabel(x)
    plt.ylabel(y)
# 이름을 셋팅해주는 함수
```

### Setosa

```python
setosa20_group.plot.hist()
setting("setosa histogram","DATA","CM")

setosa20_group.plot.hist(bins=3)
setting("setosa histogram","DATA","CM")

setosa20_group.hist()

setosa20_group.hist(bins=3)
```

### versicolor

```python
versicolor20_group.plot.hist()
setting("versicolor histogram","DATA","CM")

versicolor20_group.plot.hist(bins=3)
setting("versicolor histogram","DATA","CM")

versicolor20_group.hist()

versicolor20_group.hist(bins=3)
```

### virginca

```python
virginica20_group.plot.hist()
setting("virginica histogram","DATA","CM")

virginica20_group.plot.hist(bins=3)
setting("virginica histogram","DATA","CM")

virginica20_group.hist()

virginica20_group.hist(bins=3)
```

## 3개 종에 대한 scatter plot을 결합

### 3-1 sepal length ,petal length 사용

```python
ax = setosa20_group.plot.scatter(x='Sepal length', y='Petal length', c='Red',label='setosa')
ax = versicolor20_group.plot.scatter(x='Sepal length', y='Petal length', c='Green',label='versicolor',ax=ax)
ax = virginica20_group.plot.scatter(x='Sepal length', y='Petal length', c='Blue',label='virginica',ax=ax)
```

### 3-2 sepal width ,petal width 사용

```python
ax = setosa20_group.plot.scatter(x='Sepal width', y='Petal width', c='Red',label='setosa')
ax = versicolor20_group.plot.scatter(x='Sepal width', y='Petal width', c='Green',label='versicolor',ax=ax)
ax = virginica20_group.plot.scatter(x='Sepal width', y='Petal width', c='Blue',label='virginica',ax=ax)
```

## 결합된 데이터의 scatter matrix

```python
import mglearn
iris_class = iris20['Species'].map({"setosa": 0,"versicolor": 1,"virginica": 2})

# kde
pd.plotting.scatter_matrix(iris20,s=60,figsize=(15, 15),marker='o',c=iris_class,
                           diagonal='kde',cmap = mglearn.cm3)

# histogram
pd.plotting.scatter_matrix(iris20,s=60,figsize=(15, 15),marker='o',c=iris_class,
                           diagonal='hist',cmap = mglearn.cm3)
```

## 결합된 데이터의 평균에 대한 bar

```python
# 결합된 데이터
iris_mean = iris20.mean()

iris_mean.plot.bar(rot=0)
```

```python
# 품종별로
iris_class = iris20.groupby(iris20['Species']).mean()
iris_class.columns.name = 'feature'

iris_class.plot.bar(rot=0)
```

## 결합된 데이터의 평균에 대한 pie plot 

```python
iris_mean.plot.pie(title = "iris20 mean")
```

```python
iris_class.plot.pie(title = "Sepal length mean",y="Sepal length")
iris_class.plot.pie(title = "Sepal width mean",y="Sepal width")
iris_class.plot.pie(title = "Petal length mean",y="Petal length")
iris_class.plot.pie(title = "Petal width mean",y="Petal width")
```
