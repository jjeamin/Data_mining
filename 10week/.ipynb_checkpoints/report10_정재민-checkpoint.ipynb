# diabetes 데이터 사용

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import mglearn

diabetes = load_diabetes()
```

# 데이터 설명 (형태 특징이름)

```python
print(diabetes['DESCR'])

print("특성의 이름: {}".format(diabetes['feature_names']))
print("data의 타입: {}".format(type(diabetes['data'])))
print("data의 크기: {}".format(diabetes['data'].shape))
print("data의 처음 다섯 행:\n{}".format(diabetes['data'][:5]))
print("target의 타입: {}".format(type(diabetes['target'])))
print("타깃:\n{}".format(diabetes['target']))
```

# 데이터 프레임을 생성해서 타입,요약통계,상위 10개 확인

```python
df = pd.DataFrame(diabetes['data'],columns=diabetes['feature_names'])
df['target'] = diabetes['target']
display(df)
```

```python
print("타입")
print(type(df))
print("요약통계")
print(df.describe())
print("상위 10개")
print(df.head(10))
print("하위 10개")
print(df.tail(10))
```

# 개별 특징 분포확인 histogram

```python
df.hist(bins=10,figsize=(9,7))
```

# 당뇨결과와 다른 특징들과의 선형관계 확인

## Scatter plot

```python
for column in df.columns:
    df.plot.scatter(x='target',y=column)
```

## correlation

```python
corr = df.corr(method='pearson')
display(corr)
```

# 당뇨결과를 제외한 특징들의 boxplot

```python
df_feature = df.drop(['target'],axis=1)
df_feature.boxplot()
```

# 모든 특징을 이용한 선형회귀

```python
X_train,X_test,y_train,y_test = train_test_split(df_feature,df['target'],random_state=42)

print(X_train.shape)
print(y_train.shape)
```

```python
lr = LinearRegression().fit(X_train,y_train)

w = lr.coef_
b = lr.intercept_
print("w : ", w) #계수
print("b : ", b) #절편

print("훈련 세트 점수: {:.2f}".format(lr.score(X_train,y_train)))
print("테스트 세트 점수: {:.2f}".format(lr.score(X_test,y_test)))
```

```python
prediction = lr.predict([X_test.iloc[3]])
print("예측: {}".format(prediction))
print("실제값 : {}".format(y_test.iloc[3]))
```

# 상관도가 가장 높은 특징과의 scatter plot 과 선형회귀 직선

```python
df.plot.scatter(x='target',y='bmi')
```

```python
bmi = np.array(df.bmi.tolist()).reshape(-1,1)
tar = np.array(df.target.tolist()).reshape(-1,1)

X_train,X_test,y_train,y_test = train_test_split(bmi,tar,random_state=42)

lr = LinearRegression().fit(X_train,y_train)

print("훈련 세트 점수: {:.2f}".format(lr.score(X_train,y_train)))
print("테스트 세트 점수: {:.2f}".format(lr.score(X_test,y_test)))

df.plot.scatter(x='bmi',y='target',s = [20,20],c='orange',marker = 'o')
line = np.linspace(-0.1,0.2,100).reshape(-1,1)
plt.plot(line,lr.predict(line))
```
