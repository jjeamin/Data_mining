```python
import pandas as pd
import numpy as np
from IPython.display import display
```

# 파일 읽어오기

# 심장병 DataSet

## 속성
- age : 나이,
- sex : 성별,(1:남자,0:여자)
- cp : 가슴통증유형,(1:뭘하면 자극됨 0:그 외)
- trestbps : 혈압,
- chol : 혈청,
- fbs : 혈당(>120mg/dl)
- restecg : 휴식후 심전도 회복결과,(1:휴식 회복 0: 회복 x)
- thalach : 최대 심박수,
- exang : 협십증,
- oldpeak : ST우울증,
- slope : ST 우울증에 대한 경사도, 
- ca : flourosopy에 의해서 색칠된 주요 혈관의 수
- thal : 3 = 정상 6 = 고정결합 7 = 역전가능한 결함
- 생존유무

순서입니다. 그리고 0과 1로 이루어진 항목은 `1: 참 0: 거짓`으로 나누어 집니다.

```python
df = pd.read_csv('heart.csv')

display(df)
```

# index,columns 확인

```python
print("==index==")
print(df.index)
print("==columns==")
print(df.columns)
```

# 상위, 하위 데이터 확인

```python
# 상위
print("==상위==")
display(df.head(5))

# 하위
print("==하위==")
display(df.tail(5))
```

# 통계값

```python
# 소수점 2번째 자리까지 출력
pd.options.display.float_format = '{:.2f}'.format

count = df.count()
csum = df.sum()
rsum = df.sum(axis=1)
mean = df.mean()
min1 = df.min()
max1 = df.max()

df_stat = df.copy()
df_stat.loc['count'] = count
df_stat.loc['csum'] = csum
df_stat.loc['mean'] = mean
df_stat.loc['min'] = min1
df_stat.loc['max'] = max1
df_stat['rsum'] = rsum

display(df_stat)
```

# Nan이 속한 데이터 삭제

```python
df_nan = df.dropna(how='any')

display(df_nan)
```

# sex 속성을 변경(0 : female 1 : male)

```python
df.loc[df.sex == 1,'sex'] = 'female'
df.loc[df.sex == 0,'sex'] = 'male'

display(df)
```

# 40미만, 40-55, 50초과 연령에 대한 개수, 평균 확인 

```python
# 40대 미만
df1 = df.copy()

m = (df.age < 40)
df1 = df1[m]

df1.loc['count'] = df1.count()
df1.loc['mean'] = df1.mean()

display(df1)
```

```python
# 40~55
df2 = df.copy()

m = (df.age < 55) & (df.age >= 40)
df2 = df2[m]

df2.loc['count'] = df2.count()
df2.loc['mean'] = df2.mean()

display(df2)
```

```python
# 50 초과
df3 = df.copy()

m = (df3.age > 50)
df3 = df3[m]

df3.loc['count'] = df3.count()
df3.loc['mean'] = df3.mean()

display(df3)
```
