# irisData.csv 읽기

```python
df = pd.read_csv('irisData.csv')

display(df)
```

# 상위 하위 5줄 확인

```python
# 상위
display(df.head(5))

# 하위
display(df.tail(5))
```

# index column 살펴보기

```python
# index
print("index")
print(df.index)

# column
print("column")
print(df.columns)
```

# 통계값 확인

```python
count = df.count()
csum = df.sum()
rsum = df.sum(axis=1)
mean = df.mean()
min1 = df.min()
max1 = df.max()

df_stat = df.copy()
df_stat.loc['count'] = count
df_stat.loc['csum'] = csum
df_stat.loc['mean'] = mean
df_stat.loc['min'] = min1
df_stat.loc['max'] = max1
df_stat['rsum'] = rsum

df_stat['Species'] = df_stat['Species'][:-4] # species는 문자열이라 특정 통계값 제거

display(df_stat)
```

# Species 속성값 확인

```python
print('==Species==')
display(df['Species'])

print("==중복값 제거 하고 보기 쉽게 표현==")

# 중복값을 제거하고 어떤 종이 있는지 출력
display(df.drop_duplicates(['Species'])['Species'])
```

# 꽃잎 길이(Petal length) 역순 출력

```python
print('==Petal length==')
display(df['Petal length'][::-1])
```

# 종이 setosa만 출력

```python
mask = (df['Species'] == 'setosa')
display(df[mask])
```

# 꽃받침길이(Sepal length)가 5이하인 데이터만 추출

```python
mask2 = (df['Sepal length'] <= 5)
display(df[mask2])
```

# 폭 속성만 추출

```python
# 폭 속성
df_width = df.drop(['Sepal length','Petal length'],axis=1) # 원본에 변화가 없음

display(df_width)
```

# 길이 속성만 추출

```python
# 길이 속성
df_length = df.drop(['Sepal width','Petal width'],axis=1) # 원본에 변화가 없음

display(df_length)
```

# 꽃받침 길이( Sepal length)와 꽃받침 폭( Sepal width)의 비율을 새로운 속성으로 추가

```python
# sepal ratio
df_sepal_ratio = df.copy()

df_sepal_ratio['Sepal ratio'] = df['Sepal length'] / df['Sepal width']

display(df_sepal_ratio)
```

# 꽃잎 길이(Petal length) 와 꽃잎 폭(Petal width)의 비율을 새로운 속성으로 추가

```python
# petal ratio
df_petal_ratio = df.copy()

df_petal_ratio['Petal ratio'] = df['Petal length'] / df['Petal width']

display(df_petal_ratio)
```
