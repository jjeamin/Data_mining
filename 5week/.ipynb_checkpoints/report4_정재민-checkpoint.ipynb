```python
import pandas as pd
import numpy as np
from IPython.display import display
```

# 파일 읽어오기

```python
df = pd.read_csv('heart.csv')

display(df)
```

# index,columns 확인

```python
print("==index==")
print(df.index)
print("==columns==")
print(df.columns)
```

# 상위, 하위 데이터 확인

```python
# 상위
print("==상위==")
display(df.head(5))

# 하위
print("==하위==")
display(df.tail(5))
```

# 통계값

```python
# 소수점 2번째 자리까지 출력
pd.options.display.float_format = '{:.2f}'.format

count = df.count()
csum = df.sum()
rsum = df.sum(axis=1)
mean = df.mean()
min1 = df.min()
max1 = df.max()

df_stat = df.copy()
df_stat.loc['count'] = count
df_stat.loc['csum'] = csum
df_stat.loc['mean'] = mean
df_stat.loc['min'] = min1
df_stat.loc['max'] = max1
df_stat['rsum'] = rsum

display(df_stat)
```

```python
pd.options.display.float_format = '{:.2f}'.format
```
