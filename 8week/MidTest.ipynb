```python
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt

#(1)DataFram생성
#ageGroup(1-9),sex(1,2),height,weight,BMI,bodyFat으로 구성된 파일(physicalFitness.csv)를 읽고 데이터프레임 만들기

pf = pd.read_csv('./data/physicalFitness.csv')

#(2) 데이터 타입들, 요약통계, 상,하위 5개 데이터 확인, 행레이블, 열레이블 , 크기 확인
print("pf.dtypes = ", pf.dtypes)        #데이터 타입들
print("pf.describe()=", pf.describe())  #요약통계
print("pf.head() =", pf.head())      
print("pf.tail(3) =", pf.tail(3))
print("pf.index=", pf.index)        #행 레이블
print("pf.columns=", pf.columns)   #열 레이블 
print("pf.shape=", pf.shape)   #크기 

#(3) 키가 큰 순으로 정렬
pf.sort_values(axis=0, by='height', ascending=False)
```

```python
#(4)BMI 30이상인 비만자만  ageGroup,height,weight,bodyFat을 추출하여 파일(obese.csv)로 저장
ob = pf[pf.BMI >= 30]
display(ob.head())

ob2 = ob.loc[:, ['ageGroup','height','weight','bodyFat']]
print("ob2.shape=", ob2.shape)   #크기 
display(ob2.head())

ob2.to_csv('./data/obese.csv')
```

```python
#(5) (4)에서 추출된 데이터셋에서 나이그룹 속성으로 그룹을 분리
grouped = ob2.groupby('ageGroup')

#(6)그룹에 대한 요약 통계 
r = grouped.agg(['median', 'mean', 'std'])
print('#grouped.agg()=\n', r)
```

```python
#(7) (4)에서 추출된 데이터셋의 히스토그램 그리기
ob2.hist(bins=5)
plt.show()

#(8) (4)에서 추출된 데이터셋의  weight', 'height'에 대한 박스플롯 그리기
ob2.boxplot(['weight', 'height'])
plt.title("Box Plot")
plt.show()

#(9) (4)에서 추출된 데이터셋의  weight'와 , 'bodyFat'의 관계 그래프 그리기
ob2.plot.scatter(x='weight',  y='bodyFat',   c='Blue')  #c로 색상지정
plt.title("Scatter Plot")
plt.show()
```
