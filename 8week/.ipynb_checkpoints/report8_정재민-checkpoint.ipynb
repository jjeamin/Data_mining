## 1) ageGroup(1-9),sex(1,2),height,weight,BMI,bodyFat으로 구성된 파일(physicalFitness.csv) 를 읽고 데이터프레임 만들기

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('physicalFitness.txt',sep=" ")

display(df)
```

##  2) (1)의 데이터 셋에 대한 데이터 타입들, 요약통계, 상,하위 5개 데이터 확인, 행 레이블, 열레이블 , 크기 확인

```python
print("타입 : ",type(df))
print("통계 : ",df.describe())
print("상위 5개 : ",df.head(5))
print("하위 5개 : ",df.tail(5))
print("행레이블 : ",df.index)
print("열레이블 : ",df.columns)
print("크기 : ",df.shape)
```

## 3) 키가 큰 순으로 정렬

```python
df.sort_values("height",ascending=False)
```

## 4) BMI 30이상인 비만자만  ageGroup,height,weight,BMI,bodyFat을 추출하여 파일(obese.csv) 로 저장

```python
mask = df['BMI'] >= 30
df1 = df[mask]
df1 = df1[["ageGroup","height","weight","bodyFat"]]

df1.to_csv("obese.csv")

print(df1)
```

## 5) (4)에서 추출된 데이터셋에서 나이그룹 속성으로 그룹을 분리

```python
grouped = df1.groupby('ageGroup')
```

## 1) Boston House Prices dataset 사용 https://www.kaggle.com/ 에서 검색하여  데이터 특징 설명 및 파일 

데이터베이스의 각 행은 Boston 교외 또는 마을을 나타냅니다. 데이터는 1970 년 보스턴 표준 수도권 통계 지역 (SMSA)에서 가져온 것입니다. 속성은 다음과 같이 정의됩니다 (UCI 기계 학습 저장소 1에서 가져옴)

- CRIM : 도시 별 일인당 범죄율 
- ZN : 구역 별 주거용 토지 비율 25,000 평방 피트 이상 
- INDUS : 마을 당 비 소매업 지역의 비율 
- CHAS : 찰스 리버 더미 변수 (그렇지 않으면 관로 경계 인 경우 1, 그렇지 않은 경우 0) 
- NOx : 산화 질소 농도 (1000 만분의 1)
- RM : 주거 당 평균 객실 수 
- AGE : 1940 년 이전에 건설 된 소유주 점유 단위의 비율 
- DIS : 보스턴 고용 센터를위한 가중치가있는 거리 
- RAD : 방사형 고속도로 접근성 지수 
- TAX : 1 만달 러당 부당 이득 세율 
- PTRATIO : 마을 별 학생 - 교사 비율 
- B : 1000 (Bk - 0.63) ^ 2 여기서 Bk는 도시 별 흑인 비율
- LSTAT : 인구 감소 % 상태
- MEDV : 1000 달러짜리 주택 소유 주택의 중간 값

```python
from sklearn.datasets import load_boston
import numpy as np

names = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']

boston = load_boston()
df2 = pd.DataFrame(data=np.c_[boston['data'], boston['target']], columns=names)

display(df2)
```

##  2) 데이터의 형태(크기), 특징 이름, 상위 데이터 10개 확인

```python
print("타입 : ",type(df2))
print("크기 : ",df2.shape)
print("특징 이름 : ",df2.columns)
print("상위 : ",df2.head(10))
```

## 3) 모든 특징들의 분포 확인

```python
df2.hist(bins=10,figsize=(9,7),grid=False)
```

## 4)모든 특징들 사이의  상관계수

```python
corr=df2.corr(method='pearson')
display(corr)
```

## 5)주택가격과 다른 특징들과의 관계 확인 ( MEDV와 다른 특징과의 상관계수)

```python
corrMEDV = corr['MEDV']
display(corrMEDV)
```

## 6)주택가격과 다른 특징들에 대한 scatter plot 

```python
for column in df2.columns:
    if column == 'MEDV':
        continue
    df2.plot.scatter(x = column,y = 'MEDV',title = column)
```
